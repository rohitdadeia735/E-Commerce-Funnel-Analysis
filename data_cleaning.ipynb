```python
# NOTE: Raw data files are not included in this repository due to client confidentiality.
# To run this notebook, place the source files in a local folder and update the folder path below.
import pandas as pd
import os

folder = os.path.expanduser("~/Downloads/Funnel_Shop2core/")
print("Folder set to:", folder)
print("Files found:", os.listdir(folder))
```

```python
files = {
    "Amazon B2C Jan": {"file": "MTR_B2C-JANUARY-2026-A3NCZG73R8VJ0R.csv",          "type": "csv"},
    "Amazon B2C Feb": {"file": "MTR_B2C-FEBRUARY-2026-A3NCZG73R8VJ0R.csv",         "type": "csv"},
    "Amazon B2C Mar": {"file": "MTR_B2C-MARCH-2026-A3NCZG73R8VJ0R.csv",            "type": "csv"},
    "Amazon B2C Apr": {"file": "MTR_B2C-APRIL-2026-A3NCZG73R8VJ0R (1).csv",        "type": "csv"},
    "Amazon B2B Jan": {"file": "MTR_B2B-JANUARY-2026-A3NCZG73R8VJ0R.csv",          "type": "csv"},
    "Amazon B2B Feb": {"file": "MTR_B2B-FEBRUARY-2026-A3NCZG73R8VJ0R.csv",         "type": "csv"},
    "Amazon B2B Mar": {"file": "MTR_B2B-MARCH-2026-A3NCZG73R8VJ0R.csv",            "type": "csv"},
    "Amazon B2B Apr": {"file": "MTR_B2B-APRIL-2026-A3NCZG73R8VJ0R.csv",            "type": "csv"},
    "Flipkart Jan":   {"file": "Flip Kart Sales Report Month of Jan 2026.xlsx",     "type": "excel"},
    "Flipkart Feb":   {"file": "Flipkart Sales Report Month of Feb 2026.xlsx",      "type": "excel"},
    "Flipkart Mar":   {"file": "FlipKart Sales Report Month of March 2026.xlsx",    "type": "excel"},
    "Flipkart Apr":   {"file": "FlipKart Sales Report Month of April 2026.xlsx",    "type": "excel"},
}

for name, info in files.items():
    path = folder + info["file"]
    
    if info["type"] == "csv":
        df = pd.read_csv(path, encoding="latin1")
    else:
        df = pd.read_excel(path, sheet_name="Sales Report")
    
    missing = df.isnull().sum().sum()
    
    print(f"{name}")
    print(f"  Rows: {df.shape[0]}  |  Columns: {df.shape[1]}  |  Missing values: {missing}")
    print()
```

```python
# Check which specific columns have missing values - using Amazon B2C Jan as example
df_check = pd.read_csv(folder + "MTR_B2C-JANUARY-2026-A3NCZG73R8VJ0R.csv", encoding="latin1")

missing_by_column = df_check.isnull().sum()

# Only show columns that actually have missing values
missing_by_column = missing_by_column[missing_by_column > 0]

print("Columns with missing values in Amazon B2C Jan:")
print()
for col, count in missing_by_column.items():
    pct = round(count / len(df_check) * 100, 1)
    print(f"  {col}: {count} missing ({pct}%)")
```

```python
# ── AMAZON columns we need
amazon_cols = {
    "Invoice Number":     "invoice_number",
    "Order Id":           "order_id",
    "Transaction Type":   "transaction_type",
    "Invoice Amount":     "invoice_amount",
    "Quantity":           "quantity",
    "Item Description":   "item_description",
    "Sku":                "sku",
    "Ship To State":      "ship_to_state",
    "Fulfillment Channel":"fulfillment_type",
    "Payment Method Code":"payment_method",
    "Invoice Date":       "invoice_date",
}

# ── FLIPKART columns we need (different names, same meaning)
flipkart_cols = {
    "Order ID":                                                          "order_id",
    "Order Item ID":                                                     "invoice_number",
    "Event Sub Type":                                                    "transaction_type",
    "Final Invoice Amount (Price after discount+Shipping Charges)":      "invoice_amount",
    "Item Quantity":                                                     "quantity",
    "Product Title/Description":                                         "item_description",
    "SKU":                                                               "sku",
    "Customer's Delivery State":                                         "ship_to_state",
    "Fulfilment Type":                                                   "fulfillment_type",
    "Order Date":                                                        "invoice_date",
}

# ── FILE LIST
amazon_b2c_files = [
    ("MTR_B2C-JANUARY-2026-A3NCZG73R8VJ0R.csv",       "January"),
    ("MTR_B2C-FEBRUARY-2026-A3NCZG73R8VJ0R.csv",      "February"),
    ("MTR_B2C-MARCH-2026-A3NCZG73R8VJ0R.csv",         "March"),
    ("MTR_B2C-APRIL-2026-A3NCZG73R8VJ0R (1).csv",     "April"),
]

amazon_b2b_files = [
    ("MTR_B2B-JANUARY-2026-A3NCZG73R8VJ0R.csv",       "January"),
    ("MTR_B2B-FEBRUARY-2026-A3NCZG73R8VJ0R.csv",      "February"),
    ("MTR_B2B-MARCH-2026-A3NCZG73R8VJ0R.csv",         "March"),
    ("MTR_B2B-APRIL-2026-A3NCZG73R8VJ0R.csv",         "April"),
]

flipkart_files = [
    ("Flip Kart Sales Report Month of Jan 2026.xlsx",      "January"),
    ("Flipkart Sales Report Month of Feb 2026.xlsx",       "February"),
    ("FlipKart Sales Report Month of March 2026.xlsx",     "March"),
    ("FlipKart Sales Report Month of April 2026.xlsx",     "April"),
]

# ── EMPTY LIST to collect all dataframes
all_data = []

# ── LOAD AMAZON B2C
for filename, month in amazon_b2c_files:
    df = pd.read_csv(folder + filename, encoding="latin1")
    df = df[list(amazon_cols.keys())].copy()   # keep only needed columns
    df = df.rename(columns=amazon_cols)        # rename to standard names
    df["platform"] = "Amazon"
    df["segment"]  = "B2C"
    df["month"]    = month
    all_data.append(df)
    print(f"✓ Amazon B2C {month}: {len(df)} rows loaded")

# ── LOAD AMAZON B2B
for filename, month in amazon_b2b_files:
    df = pd.read_csv(folder + filename, encoding="latin1")
    df = df[list(amazon_cols.keys())].copy()
    df = df.rename(columns=amazon_cols)
    df["platform"] = "Amazon"
    df["segment"]  = "B2B"
    df["month"]    = month
    all_data.append(df)
    print(f"✓ Amazon B2B {month}: {len(df)} rows loaded")

# ── LOAD FLIPKART
for filename, month in flipkart_files:
    df = pd.read_excel(folder + filename, sheet_name="Sales Report")
    df.columns = df.columns.str.strip()        # remove extra spaces from column names
    df = df[list(flipkart_cols.keys())].copy()
    df = df.rename(columns=flipkart_cols)
    df["platform"]      = "Flipkart"
    df["segment"]       = "B2C"
    df["month"]         = month
    df["payment_method"] = None                # Flipkart doesn't have this column
    all_data.append(df)
    print(f"✓ Flipkart {month}: {len(df)} rows loaded")

# ── MERGE ALL INTO ONE TABLE
unified = pd.concat(all_data, ignore_index=True)

print()
print(f"Total rows: {len(unified)}")
print(f"Total columns: {len(unified.columns)}")
print()
print(unified.head())
```

```python
# Check what transaction types currently exist
print("Current transaction types:")
print(unified["transaction_type"].value_counts())
```

```python
# Map everything to one standard label
mapping = {
    "Shipment":           "Shipped",
    "Sale":               "Shipped",
    "Refund":             "Returned",
    "Return":             "Returned",
    "Cancel":             "Cancelled",
    "Cancellation":       "Cancelled",
    "FreeReplacement":    "Free Replacement",
    "Return Cancellation":"Return Cancellation",
}

unified["transaction_type"] = unified["transaction_type"].map(mapping)

# Verify
print("After normalization:")
print(unified["transaction_type"].value_counts())
```

```python
# Add month number for correct sorting
month_order = {
    "January":  1,
    "February": 2,
    "March":    3,
    "April":    4,
}

unified["month_num"] = unified["month"].map(month_order)

# Verify
print(unified[["month", "month_num"]].drop_duplicates().sort_values("month_num"))
```

```python
print("Shape:", unified.shape)
print()
print("Columns:", list(unified.columns))
print()
print("Missing values:")
print(unified.isnull().sum())
print()
print("Platform breakdown:")
print(unified.groupby(["platform", "segment", "month"])["order_id"].count())
```

```python
# Save to the same folder as your original files
output_path = folder + "unified_sales.csv"

unified.to_csv(output_path, index=False)

print(f"File saved successfully!")
print(f"Location: {output_path}")
print(f"Rows: {len(unified)}")
print(f"Columns: {len(unified.columns)}")
```
